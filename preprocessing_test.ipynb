```python

# import cv2
# import numpy as np
# import matplotlib.pyplot as plt
# from google.colab import files
# from io import BytesIO

# def upload_and_remove_horizontal_lines():
#     # Upload the image file
#     uploaded = files.upload()

#     if not uploaded:
#         print("No file was uploaded.")
#         return

#     # Get the uploaded file
#     file_name = next(iter(uploaded))
#     file_content = uploaded[file_name]

#     # Read the image from the uploaded content
#     file_bytes = np.asarray(bytearray(file_content), dtype=np.uint8)
#     image = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

#     if image is None:
#         print("Error: Image cannot be read.")
#         return

#     # Convert the image to grayscale
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#     # Apply binary thresholding with inversion
#     _, binary_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

#     # Define a horizontal kernel (adjust width based on line length)
#     horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (20, 1))

#     # Detect horizontal lines using morphological operations
#     detected_lines = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

#     # Dilate the detected lines slightly to make sure they can be removed completely
#     dilated_lines = cv2.dilate(detected_lines, horizontal_kernel, iterations=1)

#     # Subtract the detected and dilated lines from the binary image to keep the text
#     image_without_lines = cv2.bitwise_and(binary_image, cv2.bitwise_not(dilated_lines))

#     # Invert the image back to original format (black text on white background)
#     final_image = cv2.bitwise_not(image_without_lines)

#     # Display the result
#     plt.figure(figsize=(10, 10))
#     plt.imshow(final_image, cmap='gray')
#     plt.title('Image with Horizontal Lines Removed and Text Preserved')
#     plt.axis('off')
#     plt.show()

#     # Save the final image
#     output_path = 'processed_image.jpg'
#     cv2.imwrite(output_path, final_image)
#     print(f"Processed image saved as {output_path}")

#     # Offer the processed image for download
#     files.download(output_path)

# # Call the function to upload and process the image
# upload_and_remove_horizontal_lines()

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import files

def process_and_remove_horizontal_lines(image_path):
    # Read the image
    image = cv2.imread(image_path)

    if image is None:
        print(f"Error: Image {image_path} cannot be read.")
        return None

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding with inversion
    _, binary_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Define a horizontal kernel (adjust width based on line length)
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (20, 1))

    # Detect horizontal lines using morphological operations
    detected_lines = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    # Dilate the detected lines slightly to make sure they can be removed completely
    dilated_lines = cv2.dilate(detected_lines, horizontal_kernel, iterations=1)

    # Subtract the detected and dilated lines from the binary image to keep the text
    image_without_lines = cv2.bitwise_and(binary_image, cv2.bitwise_not(dilated_lines))

    # Invert the image back to original format (black text on white background)
    final_image = cv2.bitwise_not(image_without_lines)

    return final_image

def process_images_in_folder(extract_path):
    print(f"Attempting to process images in: {extract_path}")

    if not os.path.exists(extract_path):
        print(f"Error: The directory {extract_path} does not exist.")
        return

    print(f"Contents of {extract_path}:")
    print(os.listdir(extract_path))

    processed_count = 0
    for image_name in os.listdir(extract_path):
        if image_name.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
            image_path = os.path.join(extract_path, image_name)
            print(f"Processing image: {image_path}")

            processed_image = process_and_remove_horizontal_lines(image_path)

            if processed_image is not None:
                # Display the result
                plt.figure(figsize=(10, 10))
                plt.imshow(processed_image, cmap='gray')
                plt.title(f'Processed: {image_name}')
                plt.axis('off')
                plt.show()

                # Save the processed image
                output_path = f"/content/processed_{image_name}"
                cv2.imwrite(output_path, processed_image)
                print(f"Processed image saved as {output_path}")

                # Offer the processed image for download
                files.download(output_path)

                processed_count += 1
        else:
            print(f"Skipping non-image file: {image_name}")

    print(f"Processed {processed_count} images.")

# Call the function to process images from the extracted folder
extract_path = "/content/extracted_images"  # This should match the path you used in the extraction step
process_images_in_folder(extract_path)
```
